Libraries used in the code

```python
import numpy as np #https://numpy.org/
import pandas as pd #https://pandas.pydata.org/
import cv2 #https://opencv.org/
import matplotlib.pyplot as plt#https://matplotlib.org/
import os #https://docs.python.org/3/library/os.html
```

list of species needed to be resized

```python
#list of species needed to be resized
species=['Alstonia_Scholaris___healthy',
'Arjun___healthy',
'Basil___healthy',
'Chinar___healthy',
'Gauva___healthy',
'Jamun___healthy',
'Jatropha___healthy',
'Lemon___Citrus_Leaf_Miner',
'Lemon___healthy',
'Mango___healthy',
'Pomegranate___healthy',
'Pongami_Pinnata___healthy',
'Rice___BrownSpot',
'Rice___Healthy',
'Rice___Hispa',
'Rice___LeafBlast',
'wheat_healthy',
'wheat_leaf-rust']
```

```python
dirt=['val','test','train']#Data is already splited in Train ,test and val so we will take data from each folder and move it to new one after processing
```

Code for resizing images and saving the images in the perticular specie folder

```python
for k in dirt:#This loop will run from 'train','test','val'
    os.mkdir('./'+k)#first it will make the folder in the new directory of perticulara split.
    for j in species:#This loop will traverse through every folder in the species
        li=os.listdir('../input/plant-diseases-planteafx'+'/'+k+'/'+j)#it will get all the images in that perticular specie
        os.mkdir('./'+k+'/'+j)#this will make folder of that perticular specie in the new directory(train,test,val)
        for i in li:#this will triverse through every image in that perticular specie
            img1 = cv2.imread('../input/plant-diseases-planteafx'+'/'+k+'/'+j+'/'+i)#read image in the folder one by one
#             crop_img = img1[1000:3500, 500:5500]
            dim = (300, 200)#new dimentions for resized image
            resized = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA) #cv2.resize will resize the image to the new dimentions
            cv2.imwrite('./'+k+'/'+j+'/'+i,resized)#saving the resized image with the name on the perticular specie folder.
            # plt.imshow(img1)
            # plt.show()
            # plt.imshow(resized)
            # plt.show()
```
