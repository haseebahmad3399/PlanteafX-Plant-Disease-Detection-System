Following is the code to increase and decrease the brightness on the image for augmenataion

```python
for k in dirt:#This loop will run from 'train','test','val'
    os.mkdir('F:\data1\\'+k)#first it will make the folder in the new directory of perticulara split.
    for j in species:#This loop will traverse through every folder in the species
        li=os.listdir(r'F:\Data\\'+k+'\\'+j)#it will get all the images in that perticular specie
        os.mkdir('F:\data1\\'+k+'\\'+j)#this will make folder of that perticular specie in the new directory(train,test,val)
        for i in li:#this will triverse through every image in that perticular specie
            image = cv2.imread(r'F:\Data\\'+k+'\\'+j+'\\'+i)#read image in the folder one by one
            
            bright = np.ones(image.shape , dtype="uint8") * 30#np.ones will generate an array same as shape of image and fill it with one.then multiply it with 30
            brightincrease = cv2.add(image,bright)#cv2.add will add image and bright and in the result it will increase the pixel values of the image which cause in the increase in brightness of the image
            name=str(30)+'_'+i#name for the new image
            cv2.imwrite('F:\data1\\'+k+'\\'+j+'\\'+name, brightincrease)#saving the new image with the name on the perticular specie folder.

            bright = np.ones(image.shape , dtype="uint8") * 60#np.ones will generate an array same as shape of image and fill it with one.then multiply it with 60
            brightincrease = cv2.add(image,bright)#cv2.add will add image and bright and in the result it will increase the pixel values of the image which cause in the increase in brightness of the image
            name=str(60)+'_'+i#name for the new image
            cv2.imwrite('F:\data1\\'+k+'\\'+j+'\\'+name, brightincrease)#saving the new image with the name on the perticular specie folder.          

            bright = np.ones(image.shape , dtype="uint8") * 15#np.ones will generate an array same as shape of image and fill it with one.then multiply it with 15
            brightdecrease = cv2.subtract(image,bright)#cv2.subtract will subtract image and bright and in the result it will decrease the pixel values of the image which cause in the decrease in brightness of the image
            name=str(15)+'_'+i#name for the new image
            cv2.imwrite('F:\data1\\'+k+'\\'+j+'\\'+name, brightdecrease)#saving the new image with the name on the perticular specie folder.


```

Following is the code to copy images from augmented image folder to final folder (almost 2000 images each)

```python

for k in dirt:#This loop will run from 'train','test','val'
    for j in species:#This loop will traverse through every folder in the species
        li=os.listdir(r'F:\asda\\'+k+'\\'+j)#it will get all the images in that perticular specie
        length=int((len(li)-5)/4)#it will get the number of images in the augmented folder of the perticular specie and divide it with 4,this is to divide the whole images in the folder to be divided into 4 parts
        listt=[]
        lenn=len(os.listdir(r'E:\final_dataset\Dataset-first-Splitted(.80, .20) - Copy\\'+k+'\\'+j))#it will get the number of images in the actual folder to know if it needed the augmentation or not 
        while (lenn < 2000) :#if the length of the actual folder is less than 2000 then we will augment it and copy images from augmented folder.
            n1=random.randint(1,length)#it will generate a random number between 1 and length to pick an image for 1/4 of the augmented folder.
            while (n1 in listt):#check if that image is already copied or not .
                n1=random.randint(1,length)#if the image is already used to augment the it will generate a new number.

            n2=random.randint(length+1,length*2)#it will generate a random number between length+1 and length*2 to pick an image for 2/4 of the augmented folder.
            while (n2 in listt):#check if that image is already copied or not .
                n2=random.randint(length+1,length*2)#if the image is already used to augment the it will generate a new number.

            n3=random.randint((length*2)+1,(length*3))#it will generate a random number between (length*2)+1 and (length*3) to pick an image for 3/4 of the augmented folder.
            while (n3 in listt):#check if that image is already copied or not .
                n3=random.randint((length*2)+1,(length*3))#if the image is already used to augment the it will generate a new number.

            n4=random.randint((length*3)+1,(length*4))#it will generate a random number between (length*3)+1 and (length*4) to pick an image for 4/4 of the augmented folder.
            while (n4 in listt):#check if that image is already copied or not .
                n4=random.randint((length*3)+1,(length*4))#if the image is already used to augment the it will generate a new number.

            listt.append(n1)#append that image number to the list so it wont be copy again ,avoid duplicates
            listt.append(n2)#append that image number to the list so it wont be copy again ,avoid duplicates
            listt.append(n3)#append that image number to the list so it wont be copy again ,avoid duplicates
            listt.append(n4)#append that image number to the list so it wont be copy again ,avoid duplicates
            lenn+=4
            shutil.copy(r'F:\asda\\'+k+'\\'+j+'\\'+li[n1], r'E:\final_dataset\Dataset-first-Splitted(.80, .20) - Copy\\'+k+'\\'+j+'\\'+li[n1])#this will copy the image from one folder to another
            shutil.copy(r'F:\asda\\'+k+'\\'+j+'\\'+li[n2], r'E:\final_dataset\Dataset-first-Splitted(.80, .20) - Copy\\'+k+'\\'+j+'\\'+li[n2])#this will copy the image from one folder to another
            shutil.copy(r'F:\asda\\'+k+'\\'+j+'\\'+li[n3], r'E:\final_dataset\Dataset-first-Splitted(.80, .20) - Copy\\'+k+'\\'+j+'\\'+li[n3])#this will copy the image from one folder to another
            shutil.copy(r'F:\asda\\'+k+'\\'+j+'\\'+li[n4], r'E:\final_dataset\Dataset-first-Splitted(.80, .20) - Copy\\'+k+'\\'+j+'\\'+li[n4])#this will copy the image from one folder to another
            print (n1,n2,n3,n4)#prints the number of image that copied
        
```
